```python
# YOUR CODE HERE

plt.hist(data['co2'], bins=20, edgecolor = "Brown")
plt.xlabel('co2')
plt.ylabel('Frequency')
plt.title('Distribution of co2')
plt.show()
```

```python
# YOUR CODE HERE
plt.hist(data.cylinders, bins=20, edgecolor = "Brown")
plt.xlabel('cylinders')
plt.ylabel('Frequency')
plt.title('Distribution of cylinders')
plt.show()
```

```python
# YOUR CODE HERE
plt.hist(data.ghgScore, bins=20, edgecolor = "Brown")
plt.xlabel('gghScore')
plt.ylabel('Frequency')
plt.title('Distribution of ghgScore')
plt.show()
```

```python
data_num = data.select_dtypes(include=np.number)
tc = data_num.corr()
sns.heatmap(tc, annot=False,
            cmap = "YlGnBu",
            fmt=".2f",
)
plt.show()
```

There's a high/direct relationship between the number of cylinders as well as engine displacement and Co2 emmisions thus the blue color(0.75 - 1 correlation coefficient).

There's little to no relationship between model year and cylinders thus the green color.

Alternatively, there's an inverse relationship between the number of cylinders and ghgScore and fe score thus the yellow color.
